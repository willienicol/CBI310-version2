### But first some basics

To understand AD we need to understand the electron. Do you know that not all electrons are equal? Life without oxygen is very interesting. Oxygen is the big electron acceptor. But what is this electron that keeps on being mentioned? Soon we’ll start talking about inter-specie (even inter-domain) electron transfer and for this, we need to understand the electron and the proton, or when put together simply hydrogen.

<img src="hydrogen.png" width="350" />

We now have a good idea of what food is. In the microbial world, we use the word substrate. Think of glucose, amino acids, fatty acids, and other sugar monomers when you think of food/substrate. Hydrogen is present in food mostly in bonds with carbon, and to a lesser extent in bonds with oxygen and nitrogen. Within the cell, hydrogen is ‘harvested’ in the form of NADH (after Chapter 1 you hopefully understand this well). NaDH is really biological hydrogen and there is an enzyme called [hydrogenase](https://en.wikipedia.org/wiki/Hydrogenase) that can form $H_2$ from NADH. Important to understand that NADH effectively carries two protons and two electrons. Protons are just protons, but not all electrons are equal. This is because an electron is associated with a certain energy. This energy of the electron relates to the orbital the electron is in and thus it links to the type of chemical bond the electron is used for. Electrons in $H_2$ or NADH are high energy electrons, also electrons where carbon binds with hydrogen, but electrons where hydrogen bonds to oxygen (water) are low energy electrons. Just think of [oxidative phosphorylation](https://en.wikipedia.org/wiki/Oxidative_phosphorylation), where the electrons in NADH are carefully transported to oxygen in a process where energy is harvested in the form of ATP. The mechanism involves the pumping of protons, but effectively it is the loss of electron energy that supplies the proton pumping power.

In the anaerobic world, it is hard to find suitable electron acceptors that will lower the energy of the electron. Accordingly, there is a bit of an energy crisis in the anaerobic world. We have high-energy electrons all over the place, but not enough acceptors to lower the energy of the electrons in order to make ATP. It is a bit like trying to burn petrol in a nitrogen space. All the potential, but none of the release. Of course, there is something like [substrate-level phosphorylation](https://en.wikipedia.org/wiki/Substrate-level_phosphorylation) where ATP can be formed without stripping electrons of metabolic intermediates, you will remember glycolysis of chapter 1. Microbes have developed special pathways to utilize substrate-level phosphorylation, typically producing volatile fatty acids with two, three or four carbon atoms like acetic acid, propionic acid and butyric acid. Some of these organisms also form NADH from the catabolism and without any oxygen the NADH gets expelled as $H_2$ using the hydrogenase enzyme. It is also possible for the NADH electron to get transferred to another organism via inter-specie electron transfer. Bacteria and Archaea have a longstanding relationship in exchanging electrons, as we will see when we study anaerobic digestion in detail.

### What is anaerobic digestion (AD)?

Talking about electrons with different energy. You hopefully understand that high-energy electrons are a result of capturing sunshine. The carbs that get generated in photosynthesis contain the carbon-hydrogen bonds where high-energy electrons live. If we take any plant produce, whether food or non-edible we thus have high energy electrons that can be used to potentially generate ATP. But as mentioned before, this is hard in an anaerobic environment where the big electron acceptor oxygen is not available. The first ATP harvest is based on substrate-level phosphorylation. The production of lactic acid is a good example. Note in the metabolic pathway (mostly glycolysis) that net ATP is formed while 4 high energy electrons are removed per glucose in the form of 2 NADH. The lactic acid bacteria can survive on the substrate-level phosphorylation ATP and discard the 4 electrons by dumping them onto pyruvic acid to form lactic acid. Important to note that lactic acid still contains a lot of potential chemical energy. With oxygen present lactic acid can be channeled to the TCA cycle and with oxidative phosphorylation lots of ATP can be generated. 

<img src="lactic.png" width="450" />

You might wonder how the starch and sugars (carbs) in the food and food waste (substrate) are converted to glucose since glucose is the starting point for glycolysis. The answer is through hydrolysis, a breakdown process facilitated by externally excreted enzymes from the lactic acid bacteria. There is no ATP gain through hydrolysis, but it is required since the lactic acid pathway requires the monomer as a starting point. Also note that proteins need to be hydrolyzed to amino acids, while fats are hydrolyzed to fatty acids and glycerol. All these monomers and their associated breakdown processes meet up at pyruvate, a very central molecule in all metabolisms. This implies that lactic acid can also be formed from a protein or fat as starting point.

We just use lactic acid as an example, but in nature, there are numerous ‘fermentation’ bacteria that excretes short chain fatty acids (2 to 4 carbons) in their quest to find ATP via substrate-level phosphorylation. Propionic, butyric and acetic acid are common catabolic products from these bacteria. The short chain fatty acids are sometimes referred to as volatile fatty acids (VFA). The VFA-producing bacteria often produce NADH as an internal product and can employ the hydrogenase enzyme to release the NADH as $H_2$ gas. Have a look at [biological hydrogen production](https://en.wikipedia.org/wiki/Biohydrogen). Alternatively, some organisms can directly transfer the electrons of the NADH over its membrane to another organism. There are also organisms that can use hydrogen directly as substrate, as long as they have the means to use the $H_2$ (or NADH). Also note that $CO_2$ is always formed in the fermentation process since the production of biomass is associated with the formation of $CO_2$.

<img src="3cul.png" width="550" />

Once the VFA, $CO_2$, and $H_2$ are formed we have another group of bacteria, referred to as the acetogens that converts VFA to acetic acid and $CO_2$ and $H_2$, gaining ATP in the process. There is also a group of acetogens that are able to use $CO_2$ as the electron acceptor while using $H_2$ as the electron donor to produce acetic acid. Amazing that these organisms can grow biomass off a gas substrate provided enough N,P, and M are present. Once the acetogens are done with their work the initial substrate (plant or organic material) is broken down to acetic acid and $H_2$. Once we have this, it is time for Archaea in the form of methanogens to perform their little dance. Methanogens produce methane!

Given the absence of $O_2$, it is understandable that some metabolic products will contain high-energy electrons. One such product is methane. Methane is special because it escapes as a gas and can be burned in an oxygen environment to generate heat. If the decomposition of biomass occurred in an aerobic environment all carbon would have been released as $CO_2$ and the aerobic organism would have gained much more ATP from the breakdown than their anaerobic counterparts. But given the methane produced in the anaerobic breakdown, humans can now harvest a significant fraction of the high-energy electrons in a usable format. We are effectively harvesting the energy that aerobic organisms would have used as ATP by burning the methane in a controlled oxygen environment. Remember the TCA cycle is like a cold burner within a cell. Methane burning by humans is a hot process. In the end, it is all about converting the high-energy electrons to low-energy electrons by making water from oxygen. 

Once we have acetate, hydrogen, and CO2 at the right pH conditions, the methanogens can thrive. There are two main types of methanogens, acetoclastic and hydrogenotrophic. The acetoclastic methanogens use only acetate as substrate and generate methane, $CO_2$, and biomass as metabolic products. The hydrogenotrophic methanogens convert $H_2$ (or NADH from interspecies electron transfer) and $CO_2$ to methane and biomass. In this process, $CO_2$ is used as an electron acceptor. The overall process results in methane and $CO_2$ accumulation in the headspace. This gas is referred to as BIOGAS.


### The single population model

Now we have covered some theory and it is time to start with our modeling approach. To make things simple we will start with a simplified model where we assume that a single culture (X) is responsible for all the anaerobic digestion conversions. It is a crude approximation, but it is a good start to get the basics in place before we move on to fancier models.

Let's consider the overall AD process, the reaction can be given by:

The overall reaction for AD can be written as follows:

$$S→X+CH_4+CO_2+NH_3+H_2O+R$$

$S$ refers to the biomass that is fed to the fermenter (substrate), $X$ is the bacterial biomass that grows from digesting the substrate, while $R$ is the recalcitrant solid fraction that consists mainly out of lignin that cannot be digested in anaerobic conditions. The biomass ($X$) consists of various organisms performing hydrolysis, acedogenesis, acetogenesis, and methanogenesis. Note that we will only work with final products and ignore intermediates like amino acids, sugar monomers and fatty acids.


Given the equation above, you have the tools to solve for the stoichiometry given the correct specifications. We already have the 4 elemental equations. The ATP (energy) balance for this system is given by: 

$$-0.45r_s=1.8r_X+\theta$$

Note that both sides of the equation are positive. You will note that the microbial population can obtain 0.45 mol of ATP per cmol of substrate consumed. The energy requirement for growth is still 1.8 mol ATP per cmol of cells(X) formed. We'll use a maximum maintenance rate of 1 $\frac{mmmol}{g \,X \cdot h}$ for this example. 

Given the energy balance, you will still require 2 additional specifications to solve for the mass balance. In this example, we will specify the growth rate ($\mu$) for the population as:

$$\mu = 0.07\frac{Cs}{K_{\mu}+Cs}$$ 

You can use a similar Monod relationship for $\theta$

$$\theta = \theta_{max}\frac{Cs}{K_{\theta}+Cs}$$ 

The **K** values are referred to as half-saturation constants (or Monod constants) and have units of concentration. Typical values used for **K** range from 1-2 $\frac{g}{L}$ of substrate. Let's convert to a cmol concentration using 1.5 $\frac{g}{L}$. We'll use the same Monod constant for growth and maintenance in this example.

Lastly, it is important to understand that not all the substrate will get digested. There is a recalcitrant or indigestible fraction that consists mostly of lignin. In this example, 20% of the substrate on a cmol basis will not be digested, accordingly:

$$r_R=-0.2r_S$$

You can use the following molecular formalus in your mass balance:

Substrate:
$$CH_{1.7}O_{0.55}N_{0.04}$$

Microbial cells
$$CH_{1.8}O_{0.5}N_{0.08}$$

$$CH_{1.7}O_{0.75}N_{0.02}$$

\begin{align*}
\frac{dC_i}{dt} &= D(C^F_i - C_i) + r_i C_X \\

Note that we are assuming a constant volume of liquid in the equation above. At this stage it is very important to understand what D is. D is the amount of liquid volume fed daily (for AD the units are days) relative to the digester liquid volume. A $D=0.05 \frac{1}{d}$ will imply that a 100 liter digester is fed with 5 liters in a day. Some people prefer to talk about the inverse of D and call it hydraulic retention time (HRT). In this case it will be 40 days. This implies that the average inert molecule will spend 40 days before exiting the digester. **You need to be very comfortable with the concept of D and HRT!**
